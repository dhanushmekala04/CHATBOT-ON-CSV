# **Importing Required Libraries for Language Model Integration and Conversational Chains**

```python
import transformers
import textwrap
import torch
from datasets import load_dataset, Dataset
from peft import LoraConfig, AutoPeftModelForCausalLM, prepare_model_for_kbit_training, get_peft_model
from transformers import AutoModelForCausalLM, AutoTokenizer, GPTQConfig, TrainingArguments

import os
import torch
from langchain.llms import HuggingFacePipeline
from langchain.text_splitter import CharacterTextSplitter
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.llms import HuggingFaceHub
from langchain.vectorstores import Chroma
from langchain.chains import ConversationalRetrievalChain
from langchain.document_loaders import CSVLoader

```

# **Loading and Configuring the Mistral-7B Model with Quantization**

The **Mistral-7B** is a high-performance large language model with 7 billion parameters. It is designed for natural language understanding and generation tasks, offering a balance between computational efficiency and accuracy. The model supports advanced fine-tuning and quantization techniques, making it suitable for various applications like chatbots, content creation, and data-driven insights, even on resource-constrained hardware.

**Significance of Quantization**: This configuration reduces the memory footprint and speeds up the processing time, making it feasible to run large-scale models on limited hardware resources while maintaining high performance. It is ideal for scenarios where hardware constraints exist, such as when running models on cloud instances or edge devices.

```python
# Load the tokenizer for the Mistral-7B model and set pad_token to eos_token
tokenizer = AutoTokenizer.from_pretrained("TheBloke/Mistral-7B-Instruct-v0.1-GPTQ")
tokenizer.pad_token = tokenizer.eos_token

# Configure quantization to use 4-bit precision, with ExLLaMA disabled
quantization_config_loading = GPTQConfig(bits=4, disable_exllama=True, tokenizer=tokenizer)

# Load the Mistral-7B model with quantization configuration and automatic device mapping
model = AutoModelForCausalLM.from_pretrained(
                            "TheBloke/Mistral-7B-Instruct-v0.1-GPTQ",
                            quantization_config=quantization_config_loading,
                            device_map="auto"
                        )
```

# **Model Configuration and Parameter-Efficient Fine-Tuning**

**Model Configuration**
Model configuration involves setting up various parameters and functionalities to optimize the performance and efficiency of a machine learning model. This step tailors the model to specific requirements, ensuring that it runs efficiently on available hardware, uses memory effectively, and is suited for the intended task.

**Purpose of Configuration:**
Model configuration is crucial for adapting large models to run efficiently on limited hardware or for specific downstream tasks. It ensures that:

Memory Consumption is minimized.

Training Speed is enhanced through optimized memory management.

Model Adaptation is feasible with techniques like LoRA, which target only a subset of model parameters.

**Parameter-Efficient Fine-Tuning (PEFT)**

Parameter-Efficient Fine-Tuning (PEFT) refers to a set of techniques designed to fine-tune large pre-trained models with significantly fewer trainable parameters than traditional methods. This approach is particularly valuable for applications involving large-scale models, such as transformers in natural language processing (NLP) and computer vision, where full fine-tuning can be computationally expensive and require extensive resources.

**Low-Rank Adaptation (LoRA)**

Low-Rank Adaptation (LoRA) is a parameter-efficient fine-tuning technique designed to adapt large pre-trained models, particularly in the context of transformer architectures commonly used in natural language processing (NLP) and other domains. The method focuses on optimizing performance while minimizing the computational resources and time required for training.



```python
model.config.use_cache=False  # Disable caching to save memory during training
model.config.pretraining_tp=1 # Set the number of pretraining tensor parallelism to 1
model.gradient_checkpointing_enable() # Enable gradient checkpointing to save memory during backpropagation
model = prepare_model_for_kbit_training(model) # Prepare the model for k-bit training (quantization)

# Configure the LoRA (Low-Rank Adaptation) settings
peft_config = LoraConfig(
    r=16, lora_alpha=16, lora_dropout=0.05, bias="none", task_type="CAUSAL_LM", target_modules=["q_proj", "v_proj"]
)
# Apply the PEFT (Parameter-Efficient Fine-Tuning) model to the base model using the defined LoRA configuration
model = get_peft_model(model, peft_config)
```

# **Text Generation Pipeline**
The pipeline allows users to quickly set up a task-specific pipeline without delving into the details of model loading and tokenization. It's especially useful for rapid experimentation and deployment

```python
# Set up a text generation pipeline with specific configurations
text_generation_pipeline = transformers.pipeline(
    model=model,
    tokenizer=tokenizer,
    task="text-generation",
    eos_token_id=tokenizer.eos_token_id,
    pad_token_id=tokenizer.eos_token_id,
    repetition_penalty=1.1,
    return_full_text=False,
    max_new_tokens=300,
    temperature = 0.3,
    do_sample=True,
)

# Create a HuggingFacePipeline object with the configured text generation pipeline
mistral_llm = HuggingFacePipeline(pipeline=text_generation_pipeline)
```

# **Data Processing and Preparation with CSV Loader and Text Splitting**

Loading the CSV File

**Splitting the Documents into Smaller Chunks:**

Splitting documents into smaller chunks is a common preprocessing step when dealing with long texts or large datasets. This technique ensures that the text is divided into more manageable portions, making it easier for various Natural Language Processing (NLP) models and downstream tasks such as embedding generation, retrieval, or question-answering.

**Embeddings:**

Embeddings are a crucial concept in Natural Language Processing (NLP) and Machine Learning, designed to transform textual data into dense numerical vectors. These vectors capture the semantic meaning of words, phrases, or entire documents, enabling machine learning models to understand, compare, and operate on text data more effectively.

```python
from langchain_community.document_loaders.csv_loader import CSVLoader
# Load the CSV file
loader = CSVLoader(file_path='/content/10000 Sales Records.csv')
data = loader.load()

# Split the documents into smaller chunks
#separator=""
text_splitter = CharacterTextSplitter(chunk_size=1000, chunk_overlap=0)
chunked_docs  = text_splitter.split_documents(data)

# Using HuggingFace embeddings
embeddings = HuggingFaceEmbeddings()
```

# **Connecting Documents to Vector Store with FAISS**

FAISS is being used to store the document embeddings and retrieve them based on semantic similarity

**What is FAISS?**
FAISS (Facebook AI Similarity Search) is a library developed by Facebook AI Research for efficient similarity search and clustering of dense vectors. It is widely used for handling and querying large-scale collections of high-dimensional vectors, such as embeddings from NLP models. FAISS provides fast and memory-efficient operations, making it ideal for tasks like semantic search, document retrieval, and recommendation systems.

```python
from langchain.vectorstores import FAISS
db = FAISS.from_documents(chunked_docs,
                          HuggingFaceEmbeddings(model_name='sentence-transformers/all-mpnet-base-v2'))


# Connect query to FAISS index using a retriever
retriever = db.as_retriever(
    search_type="similarity",
    search_kwargs={'k': 4}
)
```

# **Creating a Conversational Retrieval Chain**
 Conversational Retrieval Chain using LangChain, combining an LLM (Large Language Model) with a retriever to handle conversational queries and maintain context. This is particularly useful in scenarios where you want to query large collections of documents or datasets interactively, such as a Q&A system or a document-based chatbot.

```python
# Create the Conversational Retrieval Chain
qa_chain = ConversationalRetrievalChain.from_llm(mistral_llm, retriever,return_source_documents=True)
```

# **Interactive Query System for CSV Data Exploration**

Interactive Query System is used to query a CSV file using a conversational interface. It utilizes a Conversational Retrieval Chain with a large language model (LLM) to understand the user’s question and find the relevant information from the CSV file

```python
#We will run an infinite loop to ask questions to LLM and retrieve answers untill the user wants to quit
import sys
chat_history = []
query = input('Prompt: ')
# Ask question related CSV file like Which country has highest green gas?
result = qa_chain.invoke({'question': query, 'chat_history': chat_history})
print('Answer: ' + result['answer'] + '\n')
chat_history.append((query, result['answer']))
```
